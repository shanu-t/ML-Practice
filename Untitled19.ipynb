```python
from sklearn.preprocessing import OneHotEncoder ,OrdinalEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False)
ode = OrdinalEncoder()
```

```python
from sklearn.compose import make_column_transformer
```

```python
city2 = make_column_transformer(
    (ohe,["Segment","Country","Product"]),
    (ode,["Discount Band","Month Name"]),
    remainder = "passthrough"
)
```

```python
city2.set_output(transform="pandas")
```

```python
city2.fit_transform(df)
```
