```python
from sklearn.compose import ColumnTransformer
```

```python
def setvalue(value):
    if(value=="5more"):
        return 5
    elif(value=="more"):
        return 2
    else:
        return int(value)
        
```

```python
for i in ['2','2.1']:
    df[i]=df[i].map(setvalue)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
X = df.drop(columns = ['2.1'])
```

```python
y = df ['2.1']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
num_pipeline = Pipeline([
    ('numerical',SimpleImputer(strategy='mean')),
    ("encoding",MinMaxScaler(feature_range=(0,1)))
])
```

```python
cat_pipeline = Pipeline([
    ("Categorical",SimpleImputer(fill_value = 'missing',strategy = 'constant')),
    ("encoding",OneHotEncoder(sparse_output=False))
])
```

```python
num_cols = X.select_dtypes(include = np.number).columns.tolist()
```

```python
num_cols
```

```python
cat_cols = X.select_dtypes(include = "object").columns.tolist()
```

```python
cat_cols
```

```python
from sklearn.compose import ColumnTransformer
processor = ColumnTransformer([
    ('numerical',num_pipeline,num_cols),
    ('categorical',cat_pipeline,cat_cols)
])
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Pipe = Pipeline([
    ('preprocessor',processor),
    ('extimator',DecisionTreeClassifier())
])
```

```python
Pipe
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=11,test_size=0.2)
```

```python
Pipe.fit(X_train,y_train)
```

```python
Pipe.score(X_test,y_test)
```
